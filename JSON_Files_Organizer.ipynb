```python
import os
import shutil

# Define the source folder and the target folder for JSON files
source_folder = 'data'
target_folder = 'json_files'

# Ensure the target folder exists
os.makedirs(target_folder, exist_ok=True)

# Function to replace path separators with underscores and ensure the filename is valid
def make_valid_filename(s):
    return s.replace(os.path.sep, "_")

# Walk through the source folder
for root, dirs, files in os.walk(source_folder):
    for file in files:
        # Check if the file is a JSON file
        if file.endswith('.json'):
            # Construct the full path of the file
            full_file_path = os.path.join(root, file)
            # Compute the relative folder path
            relative_path = os.path.relpath(root, source_folder)
            # Generate the new filename using the relative folder path and the original file name
            new_filename = make_valid_filename(relative_path + "_" + file)
            # Construct the full path for the new file location
            new_file_path = os.path.join(target_folder, new_filename)
            # Copy the file to the new location with the new name
            shutil.copy2(full_file_path, new_file_path)
            print(f"Copied and renamed: {full_file_path} -> {new_file_path}")
```
